```python
import requests
from pathlib import Path
from time import sleep


with open('data/car_url.txt', 'r') as fp:
    for url in fp:
        url = url.strip()
        filename = url.split('/')[-1]
        print("Download of {0}".format(filename))
        r = requests.get(url, allow_redirects=True)
        open("data/tmp_img/{0}".format(filename), 'wb').write(r.content)
        sleep(0.1)
        
```
